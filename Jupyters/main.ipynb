```python
import pandas as pd
import requests as req
import glob
import numpy as np

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

```

```python

# Carpeta que contiene los archivos CSV
carpeta = r'C:\Users\filokaizen\OneDrive - Tajamar\Desktop\ii\last-project\Data'

# Obtener la lista de archivos CSV en la carpeta
archivos_csv = glob.glob(carpeta + "/*.csv")

# Verificar si se encontraron archivos CSV en la carpeta
if len(archivos_csv) == 0:
    print("No se encontraron archivos CSV en la carpeta especificada.")
else:
    # Lista para almacenar los DataFrames cargados
    dfs = []

    # Bucle para cargar cada archivo CSV en un DataFrame
    for archivo in archivos_csv:
        data = pd.read_csv(archivo)
        dfs.append(data)

    # Concatenar los DataFrames en uno solo
    data = pd.concat(dfs)




```

```python

data.columns = [e.lower().strip().replace(" ", "")for e in data.columns]
# e.lower : pongo todo en minúsculas
#strip, sustituye en ambos lados 
```
